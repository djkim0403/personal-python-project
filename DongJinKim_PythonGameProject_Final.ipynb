```python
# Tic-Tac-Toe game by Dong Jin Kim, 5 May 2020


# creates reference and game board
from IPython.display import clear_output

def display_board(board):
    # print('\n'*100)   <--- use this instead of clear_output() if you're not using Jupyter Notebooks
    clear_output()
    print('Number Pad Position Reference Board')
    print(' '+'7'+' | '+'8'+' | '+'9')
    print('---|---|---')
    print(' '+'4'+' | '+'5'+' | '+'6')
    print('---|---|---')
    print(' '+'1'+' | '+'2'+' | '+'3')
    print(' ')
    print(' ')
    print(' ')
    print(' '+board[7]+' | '+board[8]+' | '+board[9])
    print('---|---|---')
    print(' '+board[4]+' | '+board[5]+' | '+board[6])
    print('---|---|---')
    print(' '+board[1]+' | '+board[2]+' | '+board[3])

    
# allows player 1 to choose their desired marker
def player_input():
    marker = ''
    
    while marker != 'X' and marker != 'O':
        marker = input('Player 1, please choose your marker (X or O): ').upper()
    
    if marker == 'X':
        return ('X','O')
    else:
        return ('O','X')
    
    
# places player's marker on the chosen board position
def place_marker(board, marker, position):
    board[position] = marker

    
# checks to see if current player won
def win_check(board, mark):
    return ((board[7] == board[8] == board[9] == mark) or
    (board[4] == board[5] == board[6] == mark) or
    (board[1] == board[2] == board[3] == mark) or
    (board[7] == board[4] == board[1] == mark) or
    (board[8] == board[5] == board[2] == mark) or
    (board[9] == board[6] == board[3] == mark) or
    (board[7] == board[5] == board[3] == mark) or
    (board[1] == board[5] == board[9] == mark))


# randomly chooses which player acts first
import random

def choose_first():
    if random.randint(1,2) == 1:
        return 'Player 1'
    else:
        return 'Player 2'

    
# checks to see if position on board is available
def space_check(board, position):
    return board[position] == ' '


# checks to see if the board is all full
def full_board_check(board):
    for i in range(1,10):
        if space_check(board, i):
                return False
    return True


# takes input from player on what board position they want to place their marker
def player_choice(board):
    position = 0
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board,position):
        try:
            position = int(input(turn + ", What's your next move? "))
        except:
            continue
    return position


# asks player if they want to play again
def replay():
    replay_check = ''
    while replay_check != 'Y' and replay_check != 'N':
        replay_check = input('Play again? (Y/N): ').upper()
        
    if replay_check == 'Y':
        return True
    else:
        return False
    
    
    
# start of the game

print("Welcome to Dong Jin Kim's Tic Tac Toe!")

while True:
    board_setup = [' ']*10
    player1_marker,player2_marker = player_input()
    turn = choose_first()
    print(turn + ' is up!')
    
    ready_to_play = ''
    while ready_to_play != 'Y' and ready_to_play != 'N':
        ready_to_play = input('Ready to start the game? (Y/N): ').upper()

    if ready_to_play == 'Y':
        game_on = True
    else:
        game_on = False
        
    while game_on:
        if turn == 'Player 1':
            display_board(board_setup)
            position = player_choice(board_setup)
            place_marker(board_setup, player1_marker, position)
            
            if win_check(board_setup, player1_marker):
                display_board(board_setup)
                print('Player 1 has won the game!')
                game_on = False
            else:
                if full_board_check(board_setup):
                    display_board(board_setup)
                    print('The game is a draw!')
                    break
                else:
                    turn = 'Player 2'
                    
        else:
            display_board(board_setup)
            position = player_choice(board_setup)
            place_marker(board_setup, player2_marker, position)

            if win_check(board_setup, player2_marker):
                display_board(board_setup)
                print('Player 2 has won the game!')
                game_on = False
            else:
                if full_board_check(board_setup):
                    display_board(board_setup)
                    print('The game is a draw!')
                    break
                else:
                    turn = 'Player 1'
                
    if not replay():
        print('Thanks for playing!')
        break
```
